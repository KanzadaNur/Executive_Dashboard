Tech-Hub Sales

```python
import pandas as pd 
techhub_sales = pd.read_csv(r"C:\Users\kanza\Downloads\Week4\TechHub_Sales_Data.csv")
techhub_products = pd.read_csv(r"C:\Users\kanza\Downloads\Week4\TechHub_Products.csv")
techhub_customers = pd.read_csv(r"C:\Users\kanza\Downloads\Week4\TechHub_Customers.csv")

print(techhub_sales.head())
print(techhub_products.head())
print(techhub_customers.head())
```

```python
is_valid_product_id = techhub_sales['product_id'].isin(techhub_products['product_id']).all()

if is_valid_product_id:
    print("All product IDs in sales data are valid.")
else:
    print("There are invalid product IDs in sales data.")
```

```python
is_valid = techhub_sales['customer_id'].isin(techhub_customers['customer_id']).all()

if is_valid:
    print("All customer IDs in sales data are valid.")
else:
    print("There are invalid customer IDs in sales data.")
```

```python
merged_techhub = pd.merge(techhub_sales, techhub_products, on='product_id')
final_techhub_data = pd.merge(merged_techhub, techhub_customers, on='customer_id')
```

```python
final_techhub_data.head()
```

```python
final_techhub_data['launch_date'] = pd.to_datetime(final_techhub_data['launch_date'])
final_techhub_data['signup_date'] = pd.to_datetime(final_techhub_data['signup_date'])
final_techhub_data['order_date'] = pd.to_datetime(final_techhub_data['order_date'])
final_techhub_data[['signup_date', 'launch_date', 'order_date']].head()
```

```python
final_techhub_data.isnull().sum()
```

```python
s = final_techhub_data.duplicated()
print(s.sum())
```

```python
final_techhub_data['revenue'].dtype
```

```python
final_techhub_data['cost_price'] = final_techhub_data['cost_price'].astype(float)
final_techhub_data['quantity'] = final_techhub_data['quantity'].astype(int)
final_techhub_data['revenue'] = final_techhub_data['revenue'].astype(float)
```

```python
final_techhub_data['profit_amount'] = final_techhub_data['revenue'] - (final_techhub_data['cost_price'] * final_techhub_data['quantity'])
final_techhub_data
```

```python
final_techhub_data['profit_margin'] = (final_techhub_data['profit_amount'] / final_techhub_data['revenue']) * 100
final_techhub_data[['revenue', 'profit_amount', 'profit_margin']].head()
```

```python
import datetime as dt
final_techhub_data['tenure_days'] = (pd.Timestamp.now() - final_techhub_data['signup_date']).dt.days
final_techhub_data[['signup_date', 'order_date', 'tenure_days']].head()
```

```python
#Customer Lifetime Value = SUM([Revenue]) per customer
final_techhub_data.groupby('customer_id')['revenue'].sum().reset_index(name='customer_lifetime_value').head()
```

```python
final_techhub_data['product_age_days'] = (final_techhub_data['order_date'] - final_techhub_data['launch_date']).dt.days
final_techhub_data[['launch_date', 'order_date', 'product_age_days']].head()
```

```python
final_techhub_data1 = final_techhub_data.sort_values(by=['customer_id', 'order_date'])
final_techhub_data1[['customer_id', 'order_date']].head(20)
```

```python
orders = final_techhub_data.groupby(['customer_id', 'order_date'])['revenue'].sum()
final_techhub_data['orders'] = orders.reset_index(drop=True)
```

```python
final_techhub_data1 = final_techhub_data.sort_values(by=['customer_id', 'order_date'])
final_techhub_data1[['customer_id', 'order_date', 'orders']].head(20)
```

```python
final_techhub_data2 = final_techhub_data.groupby('customer_id')['orders'].sum().reset_index()
final_techhub_data2[['customer_id', 'orders']].head(20)
```
